```python
import pandas as pd
pd.options.display.max_rows = 999
```

```python
df = pd.read_html("https://api.census.gov/data/2019/acs/acs5/variables.html")
```

```python
len(df)
# all of the data is stored in a 1 item list
```

```python
data = df[0].iloc[:, :-1] #skip the last row of blank stuff
```

```python
geos = pd.read_csv("geo_restricts.csv")
```

```python
geos
```

```python
len(geos)
```

```python
restricts = geos[['Table Number', 'Geography Restrictions']].copy()
```

```python
restricts
```

```python
s1 = set(data['Group'])
s2 = set(restricts['Table Number'])
```

```python
len(s1.intersection(s2))
```

```python
len(s1.difference(s2))
```

```python
merge = restricts.merge(data, how='right', left_on='Table Number', right_on='Group')
```

```python
merge
```

```python
merge.loc[merge['Geography Restrictions'].notnull()]
```

```python
s1 = set(merge['Table Number'])
s2 = set(merge['Group'])
```

```python
print(len(s1), len(s2))
```

```python
len(s1.intersection(s2))
```

```python
len(s1.difference(s2))
```
